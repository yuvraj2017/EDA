```python
# importing required libraries

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# check all available datasets in seaborn

datasets = sns.get_dataset_names()

print(datasets)
```

```python
# loading the datasets 

data2 = sns.load_dataset('titanic')
print(data2)
```

```python
# plotting histogram

sns.histplot(data2["alive"])
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data2['alive'], bins=10, kde=False)
plt.title('Survived people in Titanic')
plt.xlabel('Alive')
plt.ylabel('No. of people')
plt.show()
```

```python
# Univariate Pie Chart
plt.figure(figsize=(8, 5))
data2['class'].value_counts().plot(kind='pie', autopct='%1.2f%%')
plt.title('Pie chart for Class')
plt.ylabel('')
plt.show()
```

```python
# Univariate Box Plot
plt.figure(figsize=(8, 5))
sns.boxplot(x=data2['fare'])
plt.title('Box Plot for fare rates')
plt.xlabel('fare amount in $')
plt.show()
```

```python
# Bivariate Line Plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=data2, x='age', y='class', hue='sex', err_style=None)

plt.title('Line Plot of class wise ages (colored by Gender)')
plt.xlabel('Ages')
plt.ylabel('Passenger class')
plt.legend(title='Gender')

plt.show()
```

```python
# age vs fare

plt.scatter(data2.age, data2.fare)
```

```python
# Bivariate Scatter Plot
plt.figure(figsize=(13, 8))
sns.scatterplot(x='age', y='fare', hue='class', data=data2)
plt.title('Scatter Plot of Age vs Fare (colored by Class)')
plt.xlabel('Age')
plt.ylabel('Fare amount')
plt.legend(title='Class')
plt.show()
```

```python
# Bivariate Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='class', y='fare', data=data2)
plt.title('Class wose fare rates')
plt.xlabel('Passenger class')
plt.ylabel('Fare rates)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='class', y='age', data=data2)
plt.title('Classwise ages')
plt.xlabel('Passenger class')
plt.ylabel('Age')
plt.show()
```
