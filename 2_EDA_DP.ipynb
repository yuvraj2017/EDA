# DATA PREPROCESSING

```python
titanic1.shape
```

```python
titanic1 = titanic1.dropna(axis=1)
```

```python
titanic1.shape
```

```python
titanic1.columns
```

```python
titanic1
```

```python
x = titanic1.drop('alive', axis=1)
x_numerical = x.select_dtypes(include=['int64','float64'])
```

```python
#  checking for outliers



Q1 = x_numerical.quantile(0.25)
Q3 = x_numerical.quantile(0.75)

IQR = Q3 - Q1

l_bound = Q1 - 1.5 * IQR
u_bound = Q1 + 1.5 * IQR

outlier_filter = ~((x_numerical < l_bound) | (x_numerical > u_bound)).any(axis=1)

filtered = titanic1[outlier_filter] # data after removing outliers


y = titanic1['alive']
x = titanic1.drop('alive', axis=1)

x_numerical = x.select_dtypes(include=['int64','float64'])
x_categorical = titanic1.select_dtypes(include=['object'])

x.shape, x_numerical.shape, x_categorical.shape
```

```python
x_categorical
```

```python
#  rescaling numerical columns

titanic_scale = MinMaxScaler()

x_numerical_scaled = titanic_scale.fit_transform(x_numerical)

x_numerical_scaled = pd.DataFrame(x_numerical_scaled, columns=x_numerical.columns, index = x_numerical.index)
```

```python
 # One-hot encode category columns
x_categorical_encoded = pd.get_dummies(x_categorical, drop_first=False, dtype=int)
```

```python
x_categorical_encoded
```

```python
x = pd.concat([x_numerical_scaled,x_categorical_encoded], axis=1)
```

```python
x, x.shape
```

```python
print(x.isnull().sum())
print(y.isnull().sum())
print(x.shape)
print(y.shape)
```

```python
all = pd.concat([x,y], axis=1)
all

all.to_csv("titanic_dataset.csv", index=False)
```
