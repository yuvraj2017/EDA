```python
# importing required libraries

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# check all available datasets in seaborn

datasets = sns.get_dataset_names()
print(datasets)
```

```python
#  loading the datasets

data1 = sns.load_dataset('tips')
print(data1)
```

```python
data1.head(12).to_csv("d1.csv")
```

```python
sns.histplot(data1["day"])
```

```python
# Univariate Histogram
plt.figure(figsize=(8, 5))
sns.histplot(data1['day'], bins=20, kde=True)
plt.title('No. of orders by day')
plt.xlabel('Days of Week')
plt.ylabel('Total orders')
plt.show()
```

```python
# Univariate Pie Chart
plt.figure(figsize=(8, 5))
data1['sex'].value_counts().plot(kind='pie', autopct='%1.2f%%')
plt.title('Pie chart for Genders measure')
plt.ylabel('')
plt.show()
```

```python
# Univariate Box Plot
plt.figure(figsize=(8, 5))
sns.boxplot(x=data1['total_bill'])
plt.title('Box Plot of total bills')
plt.xlabel('Amount of Total bills')
plt.show()
```

```python
# Bivariate Line Plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=data1, x='day', y='size', hue='sex', err_style=None)

plt.title('Line Plot of orders by Day (colored by Gender)')
plt.xlabel('Weekday')
plt.ylabel('order size')
plt.legend(title='Gender')

plt.show()
```

```python
# total_bill vs tip

plt.scatter(data1.total_bill, data1.tip)
```

```python
# Bivariate Scatter Plot
plt.figure(figsize=(13, 8))
sns.scatterplot(x='total_bill', y='tip', hue='day', data=data1)
plt.title('Scatter Plot of total bill vs tip (colored by Weekdays)')
plt.xlabel('Bill Amount')
plt.ylabel('Tip Amount')
plt.legend(title='Weekdays')
plt.show()
```

```python
# Bivariate Boxplot


plt.figure(figsize=(13,8))
sns.boxplot(data=data1, x="day", y="tip")
plt.title("Weekday wise tips")
plt.xlabel("Weekdays")
plt.ylabel("Tips")

plt.show()
```

```python
# Bivariate Boxplot


plt.figure(figsize=(13,8))
sns.boxplot(data=data1, x="time", y="total_bill")
plt.title("Time wise total bill of food.")
plt.xlabel("Time")
plt.ylabel("Amount of Food-bill")

plt.show()
```
